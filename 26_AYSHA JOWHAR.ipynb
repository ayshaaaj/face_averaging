```python
import os
import numpy as np
from skimage import io


image_folder_path = r'C:\Users\yamin\Downloads\lfw_100_images\lfw_100_images\Gerhard_Schroeder'
image_final = np.zeros((250, 250, 3), dtype=np.float64)


image_total = 0
for image_name in os.listdir(image_folder_path):
    image_path = os.path.join(image_folder_path, image_name)
    image = io.imread(image_path)
    image_final += image.astype(np.float64)
    image_total += 1


average_result = image_final / image_total


output_path = 'average_face.png'
io.imsave(output_path, average_result.astype(np.uint8))
print("Averaged result saved at:", output_path)
a=io.imread(output_path)
io.imshow(a)
```
